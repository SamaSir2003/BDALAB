### Q3) Prepare the data for finding association rules. Each transaction will contain a list of item in the transaction

```python
df[0] = df[0].apply(lambda x: x.split(","))
df
```

### Q4) Use Python library mlxtend and convert the transactions into a format that can be used in the Apriori method for finding frequent itemsets.  

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(df[0]).transform(df[0])
df_ = pd.DataFrame(te_ary, columns=te.columns_)
df_
```

### Q5) Find top selling items with minimum support of 2%.

```python
apriori(df_, min_support=0.02, use_colnames=True).sort_values(by="support", ascending=False)
```

### Q6) Find all frequent itemsets with minimum support of 5%

```python
apriori(df_, min_support=0.05, use_colnames=True)
```

### Q7) Find all frequent itemsets of length 2 with minimum support of 2%

```python
temp = apriori(df_, min_support=0.02, use_colnames=True)

temp["length"] = temp["itemsets"].map(lambda x: len(list(x)))
temp.loc[temp["length"] == 2, :]
```

### Q8) Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order

```python
temp = apriori(df_, min_support=0.02, use_colnames=True)
temp_ = association_rules(temp, metric="lift", min_threshold=1)
temp_.sort_values(by="confidence", ascending=False).head(10)
```

### Q9) Find association rules with minimum support of 2% and lift of more than 1.0

```python
temp = apriori(df_, min_support=0.02, use_colnames=True)
association_rules(temp, metric="lift", min_threshold=1)
```
