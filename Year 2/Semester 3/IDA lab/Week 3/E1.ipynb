```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mtcars.csv")
df
```

### Q1) Find the car with the best mpg

```python
df[df["mpg"] == df["mpg"].max()]["model"]
```

### Q2)  Find the car with the worst mpg

```python
df[df["mpg"] == df["mpg"].min()]["model"]
```

### Q3) Find the car with the best horsepower

```python
df[df["hp"] == df["hp"].max()]["model"]
```

### Q4) Find 5 number summary of displacement

```python
df[["disp"]].describe()
```

### Q5) Find median horse power

```python
df["hp"].median()
```

### Q6)  What is average mpg for manual vs. automatic cars

```python
print("Automatic: " + str(df[df["am"] == 1]["mpg"].mean()))
print("Manual: " + str(df[df["am"]  == 0]["mpg"].mean()))
print("Hence, automatic cars have higher mpg on average")
```
